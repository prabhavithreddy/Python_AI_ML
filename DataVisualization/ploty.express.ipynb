```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
from plotly.figure_factory import create_table
import plotly.express as px
px.defaults.width = 800
px.defaults.height = 400
```

```python
gapminder = px.data.gapminder()
gapminder.head()
```

```python
canada_df = gapminder[gapminder['country']=='Canada']
canada_df.head()
```

```python
fig = px.bar(canada_df, x='year', y='pop', height=300,width=400,)
fig.show()
```

```python
fig = px.bar(canada_df, x='year', y='pop', height=300,width=400,  hover_data=['lifeExp', 'gdpPercap'], 
             color='lifeExp', labels={'pop':'Population of Canada'})
fig.show()
```

```python
fig = px.scatter(gapminder.query('year == 2007'),y='lifeExp',x='gdpPercap', height=400)
fig.show()
```

```python
fig = px.scatter(gapminder.query('year == 2007'),y='lifeExp',x='gdpPercap', height=400, color='continent')
fig.show()
```

```python
fig = px.scatter(gapminder.query('year == 2007'),y='lifeExp',x='gdpPercap', height=400, color='country')
fig.show()
```

```python
fig = px.scatter(gapminder.query('year == 2007'),
                 y='lifeExp',
                 x='gdpPercap', 
                 color='country',
                size='gdpPercap',
                size_max=20)
fig.show()
```

```python
fig = px.scatter(gapminder.query('year == 2007'),
                 y='lifeExp',
                 x='gdpPercap', 
                 color='country',
                size='gdpPercap',
                size_max=20,
                hover_name='country')
fig.show()
```

```python
fig = px.scatter(gapminder,
                 y='lifeExp',
                 x='gdpPercap', 
                 color='continent',
                size='pop',
                size_max=20,
                hover_name='country',
                facet_col='continent', width=800)
fig.show()
```

```python
fig = px.scatter(gapminder,
                 y='lifeExp',
                 x='gdpPercap', 
                 color='continent',
                size='pop',
                size_max=20,
                hover_name='country',
                facet_col='continent', width=1000,height=400, log_x=True)
fig.show()
```

```python
fig = px.scatter(gapminder, 
                 x='gdpPercap', 
                 y='lifeExp', 
                 color='continent', 
                 size='pop', 
                 size_max=40, 
                 hover_name='country', 
                 log_x=True, 
                 animation_frame='year',
                 animation_group='country', 
                 range_x=[100, 100000], 
                 range_y=[25,90])
fig.show()
```

```python
fig = px.scatter(gapminder, 
                 x='gdpPercap', 
                 y='lifeExp', 
                 color='continent', 
                 size='pop', 
                 size_max=40, 
                 hover_name='country', 
                 log_x=True, 
                 animation_frame='year',
                 animation_group='country', 
                 range_x=[100, 100000], 
                 range_y=[25,90],
                 labels=dict(pop="Population", gdpPercap="GDP Per Capita", lifeExp="Life Expectency"))
fig.show()
```

```python
fig = px.line_geo(gapminder.query('year == 2007'), locations='iso_alpha', color='continent', projection='orthographic')
fig.show()
```

```python
# create a map using choropleth
fig = px.choropleth(gapminder, locations='iso_alpha', color='lifeExp', hover_name='country', 
                    animation_frame='year', color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show() 
```
